# Credit Card Fraud Detection using KNN
Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
import warnings 
warnings.filterwarnings("ignore")
```

Step 2: Load and Explore Dataset

```python
raw_data = pd.read_csv("creditcard.csv")
raw_data.head()

```

```python
print(raw_data['Class'].value_counts())
print(raw_data.info())
```

```python
corr = raw_data.corr()["Class"].drop(["Class"])
corr.plot(kind = "barh")
```

Step 3: Feature Preparation & Scaling


```python
X = raw_data.drop("Class", axis=1)
y = raw_data["Class"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Step 4: Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

Step 5: Model Training (KNN)

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

Step 6: Model Evaluation

```python
y_pred = knn.predict(X_test)

print("Accuracy_score:",accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))


y_proba = knn.predict_proba(X_test)[:,1]

roc_auc = roc_auc_score(y_test, y_proba)
print("ROC-AUC Score:", roc_auc)

```

#### Bonus step -  Choosing the correct value of k

```python
k = 10
accuracy = np.zeros((k))
std_acc = np.zeros((k))
for n in range(1,k+1):
    knn_model = KNeighborsClassifier(n_neighbors=n).fit(X_train,y_train)
    yhat = knn_model.predict(X_test)
    accuracy[n-1] = accuracy_score(y_test,yhat)
```

```python
print(f"The best accuracy of model is {accuracy.max()} at k = {accuracy.argmax()+1} ")
```

```python
plt.plot(range(1,k+1),accuracy,"r")
plt.ylabel('Model Accuracy')
plt.xlabel('Number of Neighbors (K)')
plt.title("ACCURACY SCORE REPORT")
plt.show()
```
